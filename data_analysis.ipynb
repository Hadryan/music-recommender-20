```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, SpectralClustering
```

```python
data = pd.read_csv('output.csv', error_bad_lines=False)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
features = ['hotttness', 'loudness', 'tempo', 'key', 'mode', 'time_signature', 'duration', 'familiarity']
data = data[features]
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
scaler = RobustScaler()
data_scaled = scaler.fit_transform(data)
```

```python
scaled_data = pd.DataFrame(data=data_scaled, index=data.index, columns=data.columns)
scaled_data.describe()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(scaled_data.corr())
```
